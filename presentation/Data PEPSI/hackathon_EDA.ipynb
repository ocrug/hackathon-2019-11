# Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
import os
os.getcwd()
```

```python
df = pd.read_csv('bank.csv')
df.head()
df.describe()
df.describe(include=["object"])
```

```python
# correlation between numeric features
sns.pairplot(data = df)
# we can find that numeric data here is not-symmetric
```

```python
categorical_vars = df.describe(include=["object"]).columns
continuous_vars = df.describe().columns
categorical_vars
continuous_vars
```

```python
_ = df.hist(column=continuous_vars, figsize = (16,16))
```

```python
# Count plots of categorical variables

fig, axes = plt.subplots(4, 3, figsize=(16, 16))
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.7, hspace=0.3)

for i, ax in enumerate(axes.ravel()):
    if i > 9:
        ax.set_visible(False)
        continue
    sns.countplot(y = categorical_vars[i], data=df, ax=ax)
```

```python
sns.countplot(x = 'y', data = df)
# from the plot, we find that the binary result is unbalanced
```

```python
sns.countplot(y = 'job', data = df)
```

```python
sns.countplot(y = 'education', data = df)
```

```python
# duration-箱型图
sns.boxplot(x="poutcome", y="duration", data=df, hue = 'y', palette = 'Reds')


#sns.swarmplot(x="day", y="total_bill", data=tips,color ='k',size = 3,alpha = 0.8)
```

```python
# day-琴形图
sns.violinplot(x="month", y="day", data=df,
               hue = 'y', palette="muted", 
               split=True,  # 设置是否拆分小提琴图
               inner="quartile")
```

```python
# balance-散点图
sns.stripplot(x="loan", y="balance", hue="y",
              data=df, jitter=True,
              palette="Set2",  # 设置调色盘
              dodge=True,  # 是否拆分
             )
```
